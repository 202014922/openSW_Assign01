```python
print("두 정수로 덧셈, 뺄셈, 곱셈, 나눗셈을 수행합니다.")
while True:
  try:
    a = int(input("첫 번째 정수를 입력하세요: "))
    break

  except ValueError:
    print("정수를 입력해주세요.")

while True:
  try:
    b = int(input("0이 아닌 두 번째 정수를 입력하세요: "))
    if b == 0:
      print("두 번째 정수는 0이 아니어야 합니다.")
      continue
    break

  except ValueError:
    print("정수를 입력해주세요.")

print("덧셈:", 3*(a+b))
print("뺄셈:", a-b)
print("곱셈:", a*b)
print("나눗셈:", a/b)
```
